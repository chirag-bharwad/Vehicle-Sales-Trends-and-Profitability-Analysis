```python
import matplotlib.pyplot as plt

top10 = df['Company'].value_counts().head(10)

plt.figure(figsize=(10,5))
top10.plot(kind='bar')
plt.title('Top 10 Companies by Sales')
plt.xlabel('Company')
plt.ylabel('Number of Cars Sold')
plt.show()
```

```python
df['Transmission'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Transmission Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Body Style'].value_counts().plot(
    kind='bar'
)

plt.title('Body Style Distribution')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.show()
```

```python
df['Dealer_Region'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Sales by Dealer Region')
plt.xlabel('Region')
plt.ylabel('Cars Sold')
plt.show()
```

```python
company_revenue = df.groupby('Company')['Price ($)'].sum()

company_revenue.sort_values(
    ascending=False
).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Companies by Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
region_revenue = df.groupby('Dealer_Region')['Price ($)'].sum()

region_revenue.sort_values(ascending=False).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Revenue by Dealer Region')
plt.ylabel('Revenue ($)')
plt.show()
```

```python
avg_price = df.groupby('Company')['Price ($)'].mean()

avg_price.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top 10 Companies by Average Price')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
df['Gender'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Customer Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Annual Income'].hist(bins=30)

plt.title('Annual Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Annual Income'],
    df['Price ($)']
)

plt.xlabel('Annual Income')
plt.ylabel('Price ($)')
plt.title('Income vs Car Price')
plt.show()
```
